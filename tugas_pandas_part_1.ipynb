```python
#- Jelaskan dengan kata-kata kalian sendiri: dari hasil .describes() berapa maksimal Duration, Minimal Duration, dan Rata-Rata Duration penyewaan sepeda. Bagaimana penemuan kalian?.
# Maksimal duration adalah 722236.000000 detik 
# Minimal duration adalah 60.000000	detik
# Rata rata duration adalah 1230.910141	detik
```

```python
#- replace spasi dengan '\_' di column name pada data tersebut. menggunakan metode .replace(" ", '\_')
df.columns = df.columns.str.replace(" ", '\_')
df
```

```python
# - Ganti tipe data pada columns : Start Date, End Date, dari tipe object ke tipe datetime. Menggunakan fungsi: pd.to_datetime()
df["Start\_Date"] = pd.to_datetime(df["Start\_Date"])
df["End\_Date"] = pd.to_datetime(df["End\_Date"])
df.info()
```

```python
# - Check apakah ada missing values pada data tersebut, dan ada berapa missing value?. Gunakan .isna().sum()
df.isna().sum()
```

```python
#- Remove Missing values yang ada pada column Zip Code.
df = df.dropna()
df.isna().sum()
```

```python
# - Check apakah ada nilai duplikat pada column Trip ID. Gunakan fungsi .duplicated(). Jika ada maka drop duplicate tersebut menggunakan .drop_duplicated()
df.duplicated(subset=['Trip\_ID'])
df.drop_duplicates(subset=['Trip\_ID'])
```

```python
# - Check berapa kali masing-masing dari "Start Terminal" disebutkan pada dataset tersebut. Gunakan .value_counts()
df["Start\_Terminal"].value_counts()
```
